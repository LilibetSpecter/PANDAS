## Basic Attributes & Methods of Series

**SOME COMMON ATTRIBUTES** [Official Pandas API Document](https://pandas.pydata.org/docs/reference/api/pandas.Series.html)<br>

**Series.values:** Serinin deerlerinin Numpy dizisi (array) olarak dntrlmesini salar.

**Series.index:**  Serinin indekslerinin dntrlmesini salar.

**Series.dtype:**  Serinin veri tipini dndrr.

**Series.size:**   Serinin eleman saysn dndrr.

**Series.shape:**  Serinin boyutunu dndrr.

**Series.ndim:**   Serinin boyut saysn dndrr.

**Series.head():** Serinin ilk n adet elemann dndrr.

**Series.tail():** Serinin son n adet elemann dndrr.

**Series.sample():** Serinin random n adet elemann dndrr.

**Series.describe():** Serinin istatistiksel zetini dndrr.

**Series.sort_index:** Seriyi indexlerine gre sralar.

**Series.sort_values():** Seriyi deerlerine gre sralar.

**Series.isnull():**      Serinin her bir elemannn null (None) olup olmadn kontrol eder.

**Series.fillna():**      Null deerleri belirli bir deerle doldurur.
 
**Series.dropna():**      Null deerleri seriden karr.

**Series.isin:**        Serideki elerin deerlerde yer alp almadn kontrol eder.

## Conditional Selection

```python
df=pd.DataFrame(np.random.randn(5,4), index="a b c d e".split(), columns=["A","B","C","D"])
df
```

```python
df>0 #bir kosul vermis oluyoruz. bool degerler donuyordu
```

```python
df[df>0] #0'dan buyuk olan degerleri getirttik df icine [] yazdirarak
#hepsi true oldugu icn hespini getirdi
#ama arada nagativeler olsaydi bu sefer NaN ile doldurarak getirecekti
```

```python
df[df>0] #burda NaN varsa gormek istemiyorum sadece kosulu saglayan degerleri gormek istiyorum diye;im
df[df["A"]>0] #A sutununda kosulumu saglayanlari getir dersem, kosulumu saglayan A degerlerini alip
# diger sutunlarda kosulumu saglamayanlari da getirdi.
#cunku ben sadece A sutununda arattim bu kosulu
```

```python
df[df["A"]>0][["C"]] #burda da A sutununda kosulu saglauanlardan C sutununda olanlari getir
```

```python
df[(df["A"]>0) & (df["B"]<0)] #hem A'daki 0dan buyuk olacak, hem B'deki sifirdan kucuk olancak
#iki kosulu birden saglayanlari getiriyor.
```

```python
df[(df["A"]>0) | (df["B"]<0)] #| isareti ile or demis oldum
#bu kosullardan herhangi birini saglarsa getirsin demis oluyorum
```

## Groupby & Aggregation Methods
groupby(); bir veri setindeki belirli bir stuna gre gruplama yapmak iin kullanlr.

Bu ilev, verileri gruplara ayrmak ve bu gruplar zerinde eitli ilemler yapmak iin kullanlr.

rnein, bir veri kmesinde mterilerin isimleri, yalar ve satn aldklar rnler yer alyor olabilir. groupby() ilevi kullanarak, mterilerin satn aldklar rne gre gruplandrabilir ve her rn iin ortalama yalarn bulabilirsiniz.

Pandas'n aggregation yntemleri, belirli bir veri kmesi zerinde ilemler gerekletirerek bir sonu dndrr.

Bu yntemler, bir veri kmesindeki deerleri toplama, ortalama alma, en byk veya en kk deeri bulma gibi farkl ilemleri gerekletirir.

Pandas'n aggregation yntemlerinden bazlar unlardr:

sum(): Belirtilen stunda yer alan tm deerlerin toplamn dndrr.

mean(): Belirtilen stunda yer alan tm deerlerin ortalamasn dndrr.

median(): Gruplanm verilerin medyann hesaplar.

mode() Belirtilen stunda en ok tekrar eden deerleri dndrr.

max(): Belirtilen stunda yer alan en byk deeri dndrr.

min(): Belirtilen stunda yer alan en kk deeri dndrr.

count(): Belirtilen stunda yer alan toplam deer saysn dndrr.

std(): Belirtilen stunda yer alan deerlerin standart sapmasn dndrr.

var(): Belirtilen stunda yer alan deerlerin varyansn dndrr.

describe(): Gruplanm verilerin istatistiksel zetini hesaplar.

idxmin(): Belirtilen eksen boyunca minimum deeri ieren ilk indeksi dndrr.

idxmax(): Belirtilen eksen boyunca maksimum deeri ieren ilk indeksi dndrr.

corr(): Stunlar arasndaki korelasyonu, NA/null deerleri hari tutarak hesaplar.

Bu yntemler, bir DataFrame veya Series zerinde kullanlabilir ve istenilen stun veya satrlarda ilem yaplabilir.

Pandas'n aggregation yntemleri, verileri hzl ve etkili bir ekilde analiz etmek iin ok kullanldr.
